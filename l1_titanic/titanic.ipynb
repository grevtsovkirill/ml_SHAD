```python
#4 Какого возраста были пассажиры? 
#Посчитайте среднее и медиану возраста пассажиров. В качестве ответа приведите два числа через пробел.
age = data['Age']
#print (age[1])
#print (len(age))
#print (np.nonzero(age) )
sum_a=0
std_a=0
pas_w_a=0
pas_wo_a=0
for i in range(1, len(age)+1):
#for i in range(len(age)-2, len(age)+1):
    if np.isnan(age[i]):
        pas_wo_a+=1
        continue
    pas_w_a+=1
    sum_a+=age[i]

age_wo_nan=[]    
for i in range(1, len(age)+1):
    if np.isnan(age[i]): 
        continue
    std_a+=(age[i]-sum_a/pas_w_a)**2
    age_wo_nan.append(age[i])
    
print (sum_a/pas_w_a,np.sqrt(std_a/pas_w_a), pas_wo_a)    
#889     NaN
#890    26.0
#891    32.0

XcolMean=np.mean(age,axis=0)
XcolStd=np.std(age,axis=0)
XcolMed=np.median(age_wo_nan)
print (XcolMean, XcolStd, XcolMed)
```

```python
#5Коррелируют ли число братьев/сестер/супругов с числом родителей/детей? 
#Посчитайте корреляцию Пирсона между признаками SibSp и Parch.
SS = data['SibSp']
PC = data['Parch']

SS_Mean=np.mean(SS,axis=0)
SS_Std=np.std(SS,axis=0)
PC_Mean=np.mean(PC,axis=0)
PC_Std=np.std(PC,axis=0)
nan_ss=0
covar_SS_PC=0
for i in range(1, len(PC)+1):
    nan_ss+=1
    covar_SS_PC+=(SS[i]-SS_Mean)*(PC[i]-PC_Mean)

print ('covariance: ',covar_SS_PC/len(PC))
print ('Pearson: ',(covar_SS_PC/len(PC))/(SS_Std*PC_Std) )

print (np.corrcoef(SS,PC))
```
